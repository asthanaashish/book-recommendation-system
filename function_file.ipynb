```python
import numpy as np 
import pandas as pd
```

```python
#Refining books data
books = pd.read_csv('BX-Books.csv', sep = ";", error_bad_lines=False, encoding = 'latin-1')
```

```python
#Refining users data
users=pd.read_csv('BX-Users.csv',sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
users.head(2)
```

```python
users.rename(columns={'User-ID':'user_id', 'Location':'location', 'Age':'age'}, inplace=True)
```

```python
#Refining ratings data
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'}, inplace=True)
```

```python
ratings.head(2)
```

```python
#Number of books-2+ lac
books.shape
```

```python
#Number of users-2+ lac
users.shape
```

```python
#Total ratings-11+ lac
ratings.shape
```

```python
#Considering Only those users which have rated minimum 200 books and store in x
x=ratings['user_id'].value_counts()>200
```

```python
#Extracting the user who have rated min 200 books and storing their index in y
y=x[x].index
```

```python
y
```

```python
#This step returns only those rows which among y
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
#Merging two data frames using a columns name ISBN, this step removes the books which do not have an ISBN in dataset
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
#Grouping books on the basis of title and rating and storing in number_rating
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
#Rename rating to number of rating
number_rating.rename(columns={'rating':'number of rating'}, inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.shape
```

```python
#Removing rows where both user and book is duplicating
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
#creating pivot table
book_pivot=final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
#Algo-Nearest Neighbour-Works based on clustering, this algo will compute distances and put togther the closest.But before that the zeros must be removed hence will create a sparse matrix 
#with the help of the scientific library of python-scipy
#this avoid computing the values with 0 entry and saves computation time
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
#Now calling the algo on the sparse matrix
from sklearn.neighbors import NearestNeighbors
#creating its obj
model=NearestNeighbors(algorithm='brute')
```

```python
#MODEL TRAINING 
model.fit(book_sparse)
```

```python
##ignore,just checking
distances, suggestions = model.kneighbors(book_pivot.iloc[54, :].values.reshape(1, -1), n_neighbors=6)
```

```python
suggestions
```

```python
#Loop to print the names of book recommended by the system
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
